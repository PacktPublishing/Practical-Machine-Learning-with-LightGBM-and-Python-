# General implementation of CatBoost

### 1.	Ensure we have the correct version of SciKit-Learn installed. You should have version 0.18.1 or higher. 

```python
# confirm Sci-Kit Learn version
import sklearn
sklearn.__version__
```

### 2. Import load_breast_cancer from SciKit-Learn’s datasets module. This is a common tool to create datasets where you can compare the performance of several models.

```python
# import Iris dataset
from sklearn.datasets import load_breast_cancer
```

### 3.	Next, import the main CatBoost library and ensure the right version is installed

```python
# import CatBoost 
import catboost
catboost.__version__
```

### 4.	The final import will be accuracy_score from SciKit-Learn’s metrics module. This is the main function that will be used to call out our model evaluation metric.

```python
# import accuracy score as evaluation metric
from sklearn.metrics import accuracy_score
```

### 5.	Now it is time to initialize our variables, starting with our sample dataset. We will be calling load_breast_cancer to load the Breast Cancer dataset. From there, we will create our training and test sets for our model. 

```python
# load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# create train and test sets
X_train = X[:int(len(X)*.75)]
X_test = X[int(len(X)*.75):]
y_train = y[:int(len(y)*.75)]
y_test = y[int(len(y)*.75):]
```

### 6.	As our dataset is created, we are now ready to call the CatBoost classifier model. We will create an object for the model and then we will fit our training data to the model. This is where our CatBoost model is trained on the dataset.

```python
# define CatBoost model and fit the sample dataset
model = catboost.CatBoostClassifier()
model.fit(X_train, y_train)
```

### 7.	Since our model is now trained, we are now ready to generate predictions. The features from our test set, the “X_test” values, will be inputted into our trained model to generate test predictions. These predictions will be compared with ground truth test labels (y_test) in the next step to see how well our CatBoost model fit on the sample dataset.

```python
# get your prediction set based on training features
y_pred = model.predict(X_test)
```

### 8.	Now that we have generated predictions, it is time to evaluate the model performance by computing the accuracy score on the test predictions and ground truth labels. The output will be the probability of correct predictions. 

```python
# Accuracy score between predictions and ground truth
accuracy_score(y_test, y_pred)
```
